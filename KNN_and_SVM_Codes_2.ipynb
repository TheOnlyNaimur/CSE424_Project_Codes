```python
#KNN Code
import pandas as pd

# Load the dataset to examine its structure
file_path = '/mnt/data/cleaned_earthquakes_2023_global.csv'
data = pd.read_csv(file_path)

data.info(), data.head()
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd

# Load dataset
data = pd.read_csv('/mnt/data/cleaned_earthquakes_2023_global.csv')

# Feature selection
features = data[['latitude', 'longitude', 'depth', 'mag', 'gap', 'dmin', 'rms']]
select = data['mag_category'].astype('category')
select_encoded = select.cat.codes

# Splitting dataset
M_train, M_test, N_train, N_test = train_test_split(features, select_encoded, test_size=0.3, random_state=42)

# Scaling features
scaler = StandardScaler()
M_train_scaled = scaler.fit_transform(M_train)
M_test_scaled = scaler.transform(M_test)

# KNN Model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(M_train_scaled, N_train)
knn_predictions = knn.predict(M_test_scaled)

# Evaluation
knn_precision = precision_score(N_test, knn_predictions)
print(f"KNN Accuracy: {knn_precision}")
print("KNN Classification Report:")
print(classification_report(N_test, knn_predictions, select_names=select.cat.categories))


#SVM Code
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd

# Load dataset
data = pd.read_csv('/mnt/data/cleaned_earthquakes_2023_global.csv')

# Feature selection
features = data[['latitude', 'longitude', 'depth', 'mag', 'gap', 'dmin', 'rms']]
select = data['mag_category'].astype('category')
select_encoded = select.cat.codes

# Splitting dataset
M_train, M_test, N_train, N_test = train_test_split(features, select_encoded, test_size=0.3, random_state=42)

# Scaling features
scaler = StandardScaler()
M_train_scaled = scaler.fit_transform(M_train)
M_test_scaled = scaler.transform(M_test)

# SVM Model
svm = SVC(kernel='rbf', random_state=42)
svm.fit(M_train_scaled, N_train)
svm_predictions = svm.predict(M_test_scaled)

# Evaluation
svm_precision = precision_score(N_test, svm_predictions)
print(f"SVM Accuracy: {svm_precision}")
print("SVM Classification Report:")
print(classification_report(N_test, svm_predictions, select_names=select.cat.categories))

```
